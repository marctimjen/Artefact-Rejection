```python
device = "cpu"

batch_size = 1

train_load_file = load_whole_data(ind = [i for i in range(1, 2)], input_path = r"C:\Users\Marc\Desktop\model_data_mclass\test_model_data",
                                  input_only = True)

train_file_loader = torch.utils.data.DataLoader(train_load_file,
                                                batch_size=1,
                                                shuffle=True,
                                                num_workers=0)

for file in train_file_loader:
    
    print(file[0].shape)
    result = torch.zeros(file[0].shape[1], file[0].shape[2]) # make target data
    print(result.shape)

    # the second loader is for loading the random timed 5-mins intervals
    load_series = load_5_min_intervals(file, device, start = 0)

    series_loader = torch.utils.data.DataLoader(load_series,
                                                batch_size=batch_size,
                                                shuffle=True,
                                                num_workers=0)
    for series in series_loader:
        ind, tar, chan, cut = series
        pred = ind.view(-1)
        
        cut_end = min(cut + 60*5*200, result.shape[1])
        result[int(chan)][cut:cut_end] = pred[:cut_end-cut]

    print(result)
```

```python
device = "cpu"

batch_size = 1

train_load_file = load_whole_data(ind = [i for i in range(1, 2)], input_path = r"C:\Users\Marc\Desktop\model_data_mclass\test_model_data",
                                  input_only = True)

train_file_loader = torch.utils.data.DataLoader(train_load_file,
                                                batch_size=1,
                                                shuffle=True,
                                                num_workers=0)

for file in train_file_loader:
    
    print(file[0].shape)
    result = torch.zeros(file[0].shape[1], file[0].shape[2]) # make target data
    print(result.shape)

    # the second loader is for loading the random timed 5-mins intervals
    load_series = load_5_min_intervals(file, device, start = 0)

    series_loader = torch.utils.data.DataLoader(load_series,
                                                batch_size=batch_size,
                                                shuffle=True,
                                                num_workers=0)
    for series in series_loader:
        ind, tar, chan, cut = series
        pred = model(ind)
        
        y_pred = (pred[:, 1] >= 0.5).view(-1)
        
        cut_end = min(cut + 60*5*200, result.shape[1])
        result[int(chan)][cut:cut_end] = y_pred[:cut_end-cut]
    

    print(result)
```
