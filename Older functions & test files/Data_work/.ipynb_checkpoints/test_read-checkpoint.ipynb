```python
import csv
import json
import re
import ast
```

```python
with open("C:/Users/Marc/Desktop/v1.0.0/edf/01_tcp_ar/030/00003058/s004_2012_11_29/00003058_s004_t000.lbl", "r") as file: # read rec file
    ls = csv.reader(file, delimiter='æ')
    for rows in ls:
        l = 0
        try:
            l = rows[0]
        except:
            continue
            
        if l:
            if l[:7] == "symbols":
                m = re.match(r".+(\{.+\})", l)
                string = m.group(1).replace("'", '"')
                string = re.sub(r'(\d+)', r'"\1"', string)

                sym_dict = json.loads(string) # make dict with the keys to the artifacts
                
                print(sym_dict)
                
            elif l[:5] == "label":
                m = re.match(r".+\{(.+)(\[.+\])\}", l)
                print(m.group(1).split(", "))
                ls = m.group(2).strip('][').split(", ")
                print(ls)
                ind = ls.index('1.0')
                print(ind)
                print(sym_dict[str(ind)])
```
